```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import cross_validation, metrics
```

```python
data = pd.read_csv("train_data.txt")
labels = pd.read_csv("train_labels.txt")
```

```python
data["labels"] = labels
data
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
label_encoder = LabelEncoder()
onehot_encoder = OneHotEncoder(sparse=False)
for i in range(40):
    integer_encoded = label_encoder.fit_transform(data[data.columns[190 + i]].fillna(0).values)
    #integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    #onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
    for j in range(len(data)):
        if type(data[data.columns[190 + i]].values[j]) == str:
            data[data.columns[190 + i]].values[j] = integer_encoded[j]
    print integer_encoded
```

```python
data = data.fillna(0.0)
data.head()
```

```python
data.to_csv("encoded_train_data.txt", index = False)
```

```python
train_data, test_data, train_labels, test_labels = cross_validation.train_test_split(data[data.columns[:230]], data[data.columns[230]], 
                                                                                    test_size = 0.3,
                                                                                    random_state = 1)
```

## Ridge

```python
ridge_classifier = RidgeClassifier(random_state = 1)
ridge_classifier.fit(train_data, train_labels)
ridge_predictions = ridge_classifier.predict(test_data)
```

#### ROC AUC

```python
metrics.roc_auc_score(test_labels, ridge_predictions)
```

#### precision 

```python
metrics.precision_score(test_labels, ridge_predictions, pos_label = -1)
```

```python
metrics.precision_score(test_labels, ridge_predictions)
```

#### recall

```python
metrics.recall_score(test_labels, ridge_predictions, pos_label = -1)
```

#### cross_validation

```python
scorer = metrics.make_scorer(metrics.roc_auc_score)
cv_strategy = cross_validation.StratifiedKFold(data[data.columns[230]], n_folds= 4)
```

```python
ridge_scoring = cross_validation.cross_val_score(ridge_classifier, data[data.columns[:230]], data[data.columns[230]], scoring = scorer, cv = cv_strategy)
```

```python
print 'Ridge mean:{}, max:{}, min:{}, std:{}'.format(ridge_scoring.mean(), ridge_scoring.max(), 
                                                     ridge_scoring.min(), ridge_scoring.std())
```
