```python
# кодируем категориальные признаки train_data и test_data
label_encoder = LabelEncoder()
for i in range(40):
    integer_encoded_train = label_encoder.fit_transform(train_data[train_data.columns[191 + i]].fillna(0).values)
    integer_encoded_test = label_encoder.fit_transform(test_data[test_data.columns[191 + i]].fillna(0).values)
    for j in range(len(train_data)):
        if type(train_data[train_data.columns[191 + i]].values[j]) == str:
            train_data[train_data.columns[191 + i]].values[j] = integer_encoded_train[j]
    for j in range(len(test_data)):
        if type(test_data[test_data.columns[191 + i]].values[j]) == str:
            test_data[test_data.columns[191 + i]].values[j] = integer_encoded_test[j]
```

```python
import xgboost as xgb
```

```python
# классификатор
xgb_classifier = xgb.XGBClassifier(learning_rate=0.05, max_delta_step=1, max_depth=3, min_child_weight=20, n_estimators=500, subsample=0.1) 
xgb_classifier.fit(train_data[train_data.columns[1:231]], train_data['labels'], eval_metric='auc')
xgb_predictions = xgb_classifier.predict(test_data[test_data.columns[1:]]) 
xgb_predictions_proba = xgb_classifier.predict_proba(test_data[test_data.columns[1:]]).transpose()[1]
```

```python
# выходной файл для kaggle
df = pd.DataFrame({'ID': test_data[test_data.columns[0]].values, 'result': xgb_predictions_proba})
df.to_csv('answer1.csv', index = False)
```
