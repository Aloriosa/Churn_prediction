```python
from sklearn.utils import shuffle
import pandas as pd
import numpy as np
import math
from scipy import stats
from sklearn import model_selection
```

```python
churn_data = pd.read_csv('orange_small_churn_data.train')
churn_labels = pd.read_table('orange_small_churn_labels.train', header = -1, names = ["labels"])
```

```python
churn_data_shuffled = shuffle(churn_data, random_state=123)
churn_labels_shuffled = shuffle(churn_labels, random_state=123)
```

```python
train_data, test_data, train_labels, test_labels = model_selection.train_test_split(churn_data_shuffled, churn_labels_shuffled, 
                                                                                     test_size = 0.3)
```

```python
test_data.to_csv("hold_out_data.txt", index = False)
test_labels.to_csv("hold_out_labels.txt", index = False)
```

```python
train_data.to_csv("train_data.txt", index = False)
train_labels.to_csv("train_labels.txt", index = False)
```
